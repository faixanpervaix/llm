```python
from typing import Generator, Optional

from numpy import isin

def example_books_with_author():
    return [
        {"title": "To Kill a Mockingbird", "author": "Harper Lee"},
        {"title": "1984", "author": "George Orwell"},
        {"title": "Pride and Prejudice", "author": "Jane Austen"},
        {"title": "The Great Gatsby", "author": "F. Scott Fitzgerald"},
    ]

titles = [book.get('title') for book in example_books_with_author()]


def yeild_authors(books: list[dict[str, str]]) -> Generator[Optional[str], None, None]:
    yield from [book.get('author') for book in books if isinstance(book.get('author'), str)]

for author in yeild_authors(example_books_with_author()):
    print(author)
```
