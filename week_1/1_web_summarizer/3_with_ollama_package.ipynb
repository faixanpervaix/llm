# Working with Ollama with Ollama Package

**Using Python's `ollama` Package**

```python
#imports
import ollama
from ollama._types import ChatResponse
from typing import Optional, Union
from IPython.display import Markdown, DisplayHandle, display
```

**Constants** require for connecting with local model

```python
OLLAMA_API = "http://localhost:11434/api/chat"
HEADERS = {
    "Content-Type": "application/json"
}
MODEL = "llama3.1:8b"
```

**User Message**

```python
messages = [
    {
        "role": "system",
        "content": "You are an expoer of LLM and Business Applications. You will suggest business applications of LLMs in a concise manner."
    },
    {
        "role": "user",
        "content": "List some of the business application of LLMs"
    }
]
```

**Payload**

```python
payload: dict[str, Union[str, bool, list[dict[str,str]]]] = {
    "model": MODEL,
    "messages": messages,
    "stream": False
}
```

```python
def get_markdown_content(content: str) -> Optional[DisplayHandle]:
    """
    Returns the content formatted as Markdown.
    :return: DisplayHandle for the Markdown content.
    """
    return display(Markdown(content))
try:
    response: ChatResponse = ollama.chat(
        model=MODEL,
        messages=messages,
        stream=False
    )

    if response["message"]["content"]:
        get_markdown_content(response["message"]["content"])
    else:
        print("No content received from the model.")
except Exception as e:
    print(f"An error occurred: {e}")
```
